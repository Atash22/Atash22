```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

example_sent = "This is a sample sentence, showing off the stop words filtration."

stopwords_english = set(stopwords.words('english'))

tokenized_words = word_tokenize(example_sent)

filtered_sent = []

for w in tokenized_words:
    if w not in stopwords_english:
        filtered_sent.append(w)

print(filtered_sent)
```

```python
stop_words = set(stopwords.words('english'))

txt = "Natural language processing is an important field of artificial intelligence. It focuses on helping computers understand, interpret, and generate human language. Many companies use natural language processing to analyze customer reviews, detect spam emails, and build intelligent chatbots. As technology improves, machines are becoming better at understanding context, sentiment, and meaning in text."

token_words = word_tokenize(txt)
filtered_sentence = []

for w in token_words:
  if w not in stop_words:
    filtered_sentence.append(w)

tagged = nltk.pos_tag(filtered_sentence)

print(tagged)
```

```python
import nltk

grammar = "NP: {<DT>?<JJ>*<NN>}"
chunk_parser = nltk.RegexpParser(grammar)
```

```python
sent = "The little rabbit ate the big carrot"
tokens = nltk.pos_tag(word_tokenize(sent))
chunk_result = chunk_parser.parse(tokens)
print(chunk_result)  # use print() instead of just chunk_result
```

```python
sent1 = "The little rabbit ate the big carrot in the garden"
tokens1 = nltk.pos_tag(word_tokenize(sent1))
print("POS tags:", tokens1)

grammar_np = "NP: {<DT>?<JJ>*<NN>}"
parser_np = nltk.RegexpParser(grammar_np)
result_np = parser_np.parse(tokens1)
print("\nNP chunking result:")
print(result_np)
```

```python
grammar_full = """
    NP: {<DT>?<JJ>*<NN>}
    PP: {<IN><NP>}
    VP: {<VB.*><NP|PP>*}
"""
parser_full = nltk.RegexpParser(grammar_full)
result_full = parser_full.parse(tokens1)
print("\nFull CFG result:")
print(result_full)
```

```python
sent2 = "The rabbit quickly ate a big orange carrot"
tokens2 = nltk.pos_tag(word_tokenize(sent2))

grammar_chink = """
    NP:
        {<.*>+}
        }<VBD|RB>{
"""
parser_chink = nltk.RegexpParser(grammar_chink)
result_chink = parser_chink.parse(tokens2)
print("\nChinking result (verbs/adverbs removed):")
print(result_chink)
```

```python
sentences = [
    "A large brown dog chased the small black cat",       # multiple NPs
    "She is reading an interesting book about history",   # VP + PP
    "The scientists from NASA discovered a new planet"    # NP with PP inside
]

for s in sentences:
    tokens = nltk.pos_tag(word_tokenize(s))
    result = parser_full.parse(tokens)
    print(f"\n--- {s} ---")
    result.pretty_print()
```
